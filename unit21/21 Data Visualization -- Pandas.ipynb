## <span style="color:blue"><b>1. INTRODUCTION TO DATA VISUALIZATION</b></span>

<b><i>Data visualization</b></i>, also known as data viz or data plotting, is the graphic representation of data with the goal of making the data easier to understand.

<b><i>Data visualization</b></i> can help you ...

1. Understand you data more easily.
2. See relationships between the variables in your dataset that may be hard to see otherwise.
3. Spot unusual datapoints like outliers (i.e., datapoints that lie far outside the rest of the data.

- Data visualization involves operating on a huge amount of data and converting it into meaningful and knowledgeable visuals using various tools. 
- For visualizing data we need the best software tools to handle various types of data in structured or unstructured formats from different sources such as files, web API, databases, and many more. 
- We must choose the best visualization tools ... the tools that fulfill all of our requirements. 
- If necessary, the tool(s) selected should support interactive plots generation, connectivity to data sources, combining data sources, automatically refreshing the data, securing access to data sources, and exporting widgets. All these features allow us to make the best visuals of our data and also save time.

***

- Data visualization is one of the the most important steps in the life cycle of a data science, data analytics, or data engineering project. 
- It is more impressive, interesting and understanding when we represent our study or analysis with the help of colors and graphics. 
- Using visualization elements like graphs, charts, maps, etc., it becomes easier for clients to understand the underlying structure, trends, patterns and relationships among variables within the dataset. 
- Simply explaining the data summary and analysis using plain numbers becomes complicated for both people coming from technical and non-technical backgrounds. 
- Data visualization gives us a clear idea of what the data wants to convey to us. Data visualization helps us extract and  understand the insights that can be found in the data.

Hence, seeing a graph, chart, or other visual and graphical representations of data is more pleasant and easy for our brain to process. To read and grasp text, and then convert this into a visualization of the data that might not be entirely accurate becomes difficult and time consuming to understand for the team of decision-makers. - Therefore, instead of using Matplotlib directly, it is recommended that you use the higher-level APIs provided by Pandas and Seaborn.

When no parameters are coded for the Pandas plot() method, by default this method will create a line plot with the values of the index column on the x-axis and all of the numeric values on the y-axis.

## <span style="color:blue"><b>1. LINE PLOTS</b></span>

[**pandas.DataFrame.plot.line**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.line.html)

- A line plot is useful to see how data changes over time.
- A line plot shows the relationship between two or more columns of data, often with the x-axis used to plot changes over time.

```python
mortality = pd.read_csv('mortality.csv')

mortality.head()
```

```python
mortality_wide = mortality.pivot(index = 'Year', columns = 'Age Group', values = 'Death Rate')

mortality_wide.head()
```

```python
mortality_wide.plot.line();
```

## <span style="color:blue"><b>2. AREA PLOTS</b></span>

[**pandas.DataFrame.plot.area**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.area.html)

- An area plot is useful to see how data changes over time.
- An area plot is similar to a line plot, but the data in each column is added to the data in the previous column.

```python
mortality_wide.plot.area();
```

## <span style="color:blue"><b>3. SCATTER PLOTS</b></span>

[**pandas.DataFrame.plot.scatter**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.scatter.html)

- A scatter plot is useful for showing the the relationship between two columns of numeric data.
- A scatter plot requires both x and y parameters.

```python
mortality.plot.scatter(x = 'Year', y = 'Death Rate');
```

- Notice that in this plot you can't tell which dots represent represent which age groups, even though the death rates are plotted for all four age groups.

```python
# What if you use the mortality_wide DataFrame?

mortality_wide.plot.scatter(x = 'Year', y = 'Death Rate');
```

```python
# Seaborn improves on Pandas

import seaborn as sns

sns.scatterplot(data = mortality, x = 'Year', y = 'Death Rate', hue = 'Age Group');
```

## <span style="color:blue"><b>4. BAR PLOTS</b></span>

[**pandas.DataFrame.plot.bar**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.bar.html)<br>

[**pandas.DataFrame.plot.barh**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.barh.html)

- Bar plots can be used to plot the data in different categories (i.e., to plot categorical data).
- The purpose of a bar plot is to investigate the distribution of a single categorical variable.

```python
employees = pd.read_csv('Employee_Attrition.csv', sep = ";")

employees.head()
```

```python
JobSatisfaction_cat = {
    1: 'Low',
    2: 'Medium',
    3: 'High',
    4: 'Very High'
}

JobSatisfaction_cat
```

```python
# Transform this encodings to meaninful labels

employees['JobSatisfaction'] = employees['JobSatisfaction'].map(JobSatisfaction_cat)
employees['JobSatisfaction'].head()
```

```python
employees['MaritalStatus'].value_counts().plot.bar();
```

```python
employees['MaritalStatus'].value_counts().plot.barh();
```

```python
employees['JobSatisfaction'].value_counts().plot.bar();
```

```python
employees['JobSatisfaction'].value_counts().plot.bar();
```

## <span style="color:blue"><b>5. HISTOGRAMS</b></span>

[**pandas.DataFrame.hist**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.hist.html)

- A histogram shows the frequency of the datapoints in a numeric column by diving the data on the x-axis into bins.
- By default, the Pandas plot() method uses 10 bins, but you can use the bins parameter to change that.
- Histograms will often give you a quick view of the range of the data that you are dealing with.

## <span style="color:blue"><b>6. DENSITY PLOTS</b></span>

[**pandas.DataFrame.plot.density**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.density.html)<br>

[**pandas.DataFrame.plot.kde**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.kde.html)

- Like a histogram, a density plot or a KDE (Kernal Density Plot) shows the distribution of data in a numeric column.  However, it uses a kernal density estimate (KDE) to smooth the data.
- Its y-axis represents the probability of a datapoint occurring, and that value is derived from the KDE.
- The KDE is calculated by weighting the distances of the datapoints from each location on the density line. So the more datapoints there are in a location, the higher the KDE -- and that indicates that the probability of seeing a datapoint in that location is higher.

```python
mortality.plot.density(y = 'Death Rate');
```

- Note:&emsp; To interpret -- think in terms of the probability of the next Death Rate being equal to an x-axis value.

```python
employees.plot.density(y = 'MonthlyIncome');
```

## <span style="color:blue"><b>7. BOX PLOTS</b></span>

[**pandas.DataFrame.plot.box**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.plot.box.html)<br>

[**pandas.DataFrame.boxplot**](https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.boxplot.html)

- A box plot (also known as a box-and-whisker plot) provides a quick way to get a five-number summary of the data that is being plotted.

**For a vertical box plot:**
- In a box plot, the ends of the box represent the first (lower) and third (upper) quartiles, and the horizontal line through the box represents the median or second quartile.
- The whisker at the bottom of the box goes down to the minimum value that isn't an outlier and the whisker at the top of the box goes up to the maximum value that isn't an outlier.
- Any circles above or below the whiskers represent outliers.

Note: A horizontal box plot will basically work the same way.


```python
mortality_wide.head()
```

```python
mortality_wide.plot.box();
```

```python
mortality_wide.plot.box(vert = False);
```

```python
employees[['MonthlyIncome']].plot.box();
```

```python
employees['MonthlyIncome'].plot.box(vert = False);
```

```python
employees[['Age', 'YearsAtCompany']].plot.box(vert = False);
```
